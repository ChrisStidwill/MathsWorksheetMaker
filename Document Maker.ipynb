```python
def MakeMe():
    WaitTime = 1
    Doctype = str(input("Type Q to make a quiz, S for a starter pack. "))
    if Doctype != 'Q' and Doctype != 'S' and Doctype != 'q' and Doctype != 's':
        print('Neither Q or S typed in.')
        time.sleep(WaitTime)
        MakeMe()
    if Doctype == 'S' or Doctype == 's':
        print('You have selected to make a starter pack.')
        MakeMeStarterPack(WaitTime)
        #Here, call the starter pack function
        
    elif Doctype == 'Q' or Doctype == 'q':
        print('You have selected to make a quiz.')
        #Here, call the quiz function. 

def MakeMeStarterPack(WaitTime):
    time.sleep(WaitTime)
    NumberOfTopics = int(input("How many topics do you want to include?"))
    NumericTopicArr = []
    time.sleep(WaitTime)
    print("Topics are as follows. Please type in one number at a time & press enter after typing.")
    time.sleep(WaitTime)
    for x in range(len(IQ.ITopics)):
        print(f"{x}: {IQ.ITopics[x]}")
    for x in range(NumberOfTopics):
        NumericTopicArr.append(int(input()))
    time.sleep(WaitTime)
    print(f"You've selected {NumericTopicArr}")
    time.sleep(WaitTime)
    print("Now to make this we need some prior knowledge (7 weeks' worth).")
    time.sleep(WaitTime)
    print("Input 7 topics of prior knowledge.")
    PK = []
    for x in range(7):
        PK.append(int(input()))
    NumericTopicArr = PK + NumericTopicArr
    TopicArr = []
    for x in NumericTopicArr:
        TopicArr.append(IQ.ITopics[x])
    time.sleep(WaitTime)
    Title = input("What do you want to title the document as? ")
    time.sleep(WaitTime)
    Pages = int(input("How many pages do you want per week? "))
    IQ.MakeStarterPack(Title,TopicArr,Pages)
    
    
        
    
```
