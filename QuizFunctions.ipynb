```python
import random

def MakeAddQuestion(a,b):
    c = a+b
    chance = random.randint(1,20)
    if chance == 1:
        return [f'{c} - {b} =',f'{a}']
    elif chance <= 4:
        return [f'{a} + __ = {c}',f'{b}']
    elif chance <= 7:
        return [f'__ + {a} = {c}',f'{b}']
    elif chance <= 12:
        return [f'{c} - {a} =',f'{b}']
    elif random.choice([True,False]):
        return [f'{a} + {b} =',f'{c}']
    else:
        return [f'{b} + {a} =',f'{c}']

def MakeMultQuestion(a,b):
    c = a*b
    chance = random.randint(1,20)
    if chance == 1:
        return [f'{c} ÷ {a} =',f'{b}']
    elif chance <= 4:
        return [f'{a} × __ = {c}',f'{b}']
    elif chance <= 7:
        return [f'__ × {a} = {c}',f'{b}']
    elif chance <= 12:
        return [f'{c} ÷ {a} =',f'{b}']
    elif random.choice([True,False]):
        return [f'{a} × {b} =',f'{c}']
    else:
        return [f'{b} × {a} =',f'{c}']
    
def NumericArgs(string):
    i = 0
    bNumeric = False
    NumericArg = 0
    while i<len(string):
        if string[i].isnumeric() or string[i] == '.':
            if bNumeric==False:
                NumericArg += 1
                bNumeric = True
        else:
            bNumeric = False
        i += 1
    return NumericArg

def SplitEquals(string):
    if '=' not in string:
        print('Error. SplitEquals called on a string that does not contain =')
    index = string.find('=')
    string1 = string[:index]
    string2 = string[index+1:]
    return [string1,string2]
                

def HandleNegatives(string):
    stringlocal = str(string)
    #Checks if there are 2 or more numeric arguments. Also checks its not part of some ordering question.
    if NumericArgs(stringlocal)<2 or ',' in stringlocal:
        return string
    #Performs the previous check on both sides of the = sign.
    if '=' in stringlocal:
        strings = SplitEquals(stringlocal)
        if NumericArgs(strings[0])>1:
            strings[0] = AddBrackets(strings[0])
        if NumericArgs(strings[0])>1:
            strings[1] = AddBrackets(strings[1])
        return strings[0] + '=' + strings[1]
    return AddBrackets(string)
            
    
def AddBrackets(string):
    string = ' ' + str(string)
    i = 1
    while i<len(string):
        if string[i] == '-' and string[i-1] != '(' and string[i+1].isnumeric():
            j = i + 2
            bFound = False
            while j < len(string):
                if string[j].isnumeric() or string[j]=='.':
                    j+= 1
                else:
                    bFound = True
                    break
            if bFound:
                string = string[:i] + '(' + string[i:j] + ')' + string[j:]
            else:
                string = string[:i] + '(' + string[i:] + ')'
        i += 1
    return string[1:]
            
def HandleNegativesQuiz(Quiz):
    for x in range(len(Quiz)):
        for i in range(len(Quiz[x])):
            for j in range(len(Quiz[x][i])):
                Quiz[x][i][j] = HandleNegatives(Quiz[x][i][j])


    
    
            
      
```
